# Change the tick values and labels on a colour bar

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import data as Pandas dataframe and assign columns to objects
tvz = pd.read_excel('../Data/TVZ.xlsx', sheet_name='Data')
x = tvz.EffectivePorosity_VolPercent
y = tvz.SampleDepth_mMD
z = tvz.Vp_mps

fig, ax = plt.subplots(1,1,figsize=(6,3))

# note the object assignment "cax" and "cbar"
# and then the use of the object in the next step

# make the colour scaled axis (plot)
cax = ax.scatter(x, y, c=z)

# make the colour bar with specified ticks
cbar = fig.colorbar(cax, ticks=[2000,3000,4000])

# change the tick labels
cbar.ax.set_yticklabels([2,3,4])

```
