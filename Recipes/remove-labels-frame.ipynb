# Simple Figures

Sometimes it's useful to strip elements to generate a simpler plot for a presentation or as a base of a composite figure.

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

# Note that ; are used only to suppress text output
```

```python
# Base plot

x = np.linspace(-10, 10, 100)
y = stats.norm(0,2).pdf(x)

fig, ax = plt.subplots(1,1, figsize=(5,3))
ax.plot(x, y, color='k');
```

```python
# Remove tick labels and ticks

x = np.linspace(-10, 10, 100) 
y = stats.norm(0,2).pdf(x)

fig, ax = plt.subplots(1,1, figsize=(5,3))

ax.plot(x, y, color='k')

ax.set_yticklabels([])
ax.set_xticklabels([])

ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none');
# argument can be 'left', 'right', 'both', 'default', 'none'
```

```python
# Remove axis and frame

x = np.linspace(-10, 10, 100) 
y = stats.norm(0,2).pdf(x)

fig, ax = plt.subplots(1,1, figsize=(5,3))

ax.plot(x, y, color='k')

ax.axis('off');
```
