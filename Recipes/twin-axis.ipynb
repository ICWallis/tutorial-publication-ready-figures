# Twin Axis

Add another set of axis across on the top or right of an x/y plot. 

Two examples are given:
1. Twin axis that show a different unit system for the same data on the origional axis
2. Add two sets of data to a plot that need separate axis

```python
import matplotlib.pyplot as plt
```

## Example 1 - Same data, different units

This method enables us to plot data in both metric and imperial units. 

```python
fig, ax = plt.subplots(1,1, figsize = (5, 5))

# Data
depth = [100, 700, 1500, 2400]
temp = [90, 180, 200, 200]

# Plot the data
ax.plot(
    temp, 
    depth, 
    color='k',
    marker="o",
    label="Some data" 
)

# Show legend
ax.legend(loc=3)

# Make twin axis for imperial units
ax_x = ax.twiny()
ax_y = ax.twinx()

# Define axis limits
min_depth_m = 0
max_depth_m = 2500

min_temp_degC = 0
max_temp_degC = 250

# Calculate twin limits
min_depth_ft = min_depth_m * 3.28084
max_depth_ft = max_depth_m * 3.28084

min_temp_degF = min_temp_degC / 5 * 9 + 32
max_temp_degF = max_temp_degC / 5 * 9 + 32

# Apply limits to axis
ax.set_ylim(max_depth_m,min_depth_m)
ax_y.set_ylim(max_depth_ft,min_depth_ft)

ax.set_xlim(min_temp_degC,max_temp_degC)
ax_x.set_xlim(min_depth_ft,max_temp_degF)

# Label axis
ax.set_ylabel('Depth [$m-MD$]')
ax_y.set_ylabel('Depth [$ft-MD$]')

ax.set_xlabel('Temperature [$^{\circ}C$]')
ax_x.set_xlabel('Temperature [$^{\circ}F$]')

# apply grid to twin axis
for ax in [ax_x,ax_y]:
    ax.grid(alpha=0.3)
```

## Example 2 - Different data

This method enables us to plot two sets of data that share a variable same plot. In this example, both datasets share the x axis.

Two versions of this example are given to illustrate how the legend is associated with the axis. If you want values from a twin axis to plot together in the legend with the values on the origional axis, then use the second example where a dummy is added to achieve this.

```python
# Base version

fig, ax0 = plt.subplots(1,1, figsize=(5,5))

# Data
shared = [100, 700, 1500, 2400]
big_values = [1000, 1400, 1500, 1400] 
small_values = [2, 4, 5, 6]

# Plot data on origional axis
ax0.plot(
    shared, 
    big_values, 
    color='tab:blue',
    marker="o",
    label="Big values" 
)

# Make twin axis
ax1 = ax0.twinx()

# Plot data on twin axis
ax1.plot(
    shared, 
    small_values,
    color='tab:green',
    marker="o",
    label="Small values" 
)

# Plot legend
ax0.legend(loc=2)
ax1.legend(loc=4)

# Label axis
ax0.set_xlabel('Shared axis')
ax0.set_ylabel('Origional axis (ax0)')
ax1.set_ylabel('Twin axis (ax1)')
```

```python
# Variant showing how to combine the legends

fig, ax0 = plt.subplots(1,1, figsize=(5,5))

# Data
shared = [100, 700, 1500, 2400]
big_values = [1000, 1400, 1500, 1400] 
small_values = [2, 4, 5, 6]

# Plot data on origional axis
ax0.plot(
    shared, 
    big_values, 
    color='tab:blue',
    marker="o",
    label="Big values" 
)

# Make twin axis
ax1 = ax0.twinx()

# Plot data on twin axis
ax1.plot(
    shared, 
    small_values,
    color='tab:green',
    marker="o",
    label="Small values" 
)

# Make a dummy version of small values
# so it can be included in the ax0 legend
ax0.plot(
    [], # empty list
    [],
    color='tab:green',
    marker="o",
    label="Small values" 
)

# Plot legend
ax0.legend(loc=2)
#ax1.legend(loc=4)

# Label axis
ax0.set_xlabel('Shared axis')
ax0.set_ylabel('Origional axis (ax0)')
ax1.set_ylabel('Twin axis (ax1)')
```
