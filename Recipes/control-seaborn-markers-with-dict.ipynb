```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# import excel data as a Pandas dataframe
tvz = pd.read_excel('../Data/TVZ.xlsx', sheet_name='Data')
tvz.info()
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,3))

# color dictionary defined for the rock type
color_dict = dict({
    'A':'#4675b7',  # New light blue tint off deep blue
    'A-bx':'#1953A6', # deep blue
    'A-l-bx':'#1953A6', # deep blue
    'R-i-w':'#FFA626', # deep orange 
    'R-i':'k', # black
    'T-p':'#800D80', # deep purple
    'S':'#178200', # dark green
    })

# symbol dictionary defined for the alteration assemblages
marker_dict = dict({
    'argillic, smectitie dominant':'<', 
    'argillic, illite-smectitie':'>',
    'propylitic':'s', 
    'phyllic':'o', 
    })

# plot the data using the dictionaries above
sns.scatterplot(
    x='EffectivePorosity_VolPercent',
    y='SampleDepth_mMD',
    data=tvz,
    ax=ax,
    s=90, 
    hue='RockCode', 
    palette=color_dict,
    style='AlterationAssemblage',
    markers=marker_dict,
    zorder=2, # places the plotted symbols above the grid
    )

ax.set_xlim(0,60)
ax.set_ylim(3500,0)

ax.set_xlabel('Porosity [v/v]')
ax.set_ylabel('Depth [m-VD]')

plt.minorticks_on()
ax.grid(which='major')
ax.grid(which='minor',linestyle=':')

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1.05, 0.45), ncol=1);
```
